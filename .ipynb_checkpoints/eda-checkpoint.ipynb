```python
def load_data() -> pl.LazyFrame:
    from sklearn.datasets import load_iris
    data = load_iris()
    feature_names = [name.replace('(cm)', '').strip().replace(' ', '_') for name in data['feature_names']]
    feature_names.append('target')
    schema = pl.Schema({name: pl.Float32 for name in feature_names})
    data = np.concatenate((data.data, data['target'].reshape((-1, 1))), axis=1)
    return pl.LazyFrame(data, schema)
```

```python
lf = load_data()
lf.limit(10).collect()
```

```python
sns.pairplot(lf.collect().to_pandas(), hue='target')
```

Посчитаем коэффициент корреляции Пирсона

$$r_{xy} = \frac{\sum_i^m {\left( x_i - \bar{x} \right) \left( y_i - \bar{y} \right)}} {\sqrt{\sum_i^m \left( x_i - \bar{x}\right)^2 \sum_i^m \left( y_i -\bar{y} \right)^2}} = \frac{ cov{\left( x,y \right)}} {\sqrt{s^2_x s^2_y}}$$

```python
def r_xy(x: np.ndarray, y: np.ndarray) -> float:
    return ((x - x.mean()) * (y - y.mean())).sum() / np.sqrt(((x - x.mean())**2).sum() * ((y - y.mean())**2).sum())
```

```python
r_xy(data[:, 0], data[:, 1])
```

```python
np.corrcoef(data[:, 0], data[:, 1])
```

```python
corr_matrix = np.zeros((data.shape[1], data.shape[1]))
for i in range(data.shape[1]):
    for j in range(data.shape[1]):
        corr_matrix[i, j] = r_xy(data[:, i], data[:, j])
print(corr_matrix)
```
